```python
from matplotlib import pyplot as plt
import numpy as np

x = np.random.rand(1);
x_list = []

for i in range(100):
    x = 1 + 1/x
    x_list.append(x)
#    print(x)
    
fig = plt.figure('oneOverX')
plt.plot(x_list)
plt.show()
```

```python
x_vals = np.zeros([1000,1000])

xes = np.linspace(0.001,3.999,1000)

for j in range(len(xes)):
    x = xes[j]
    for i in range(1000):
        x = np.sqrt(2)**x
        x_vals[i,j] = x
    
fig = plt.figure('sqareRoots')
plt.plot(x_vals)
plt.show()
```

```python
time = np.linspace(0,100,1000)

x_sqrt_pow = np.array([np.sqrt(1.1)**x for x in time])
x_linear = np.array([x for x in time])

y = np.transpose(np.vstack((x_sqrt_pow,x_linear)))

fig = plt.figure('sqrtPow')
plt.plot(time,y)
plt.show()
```
